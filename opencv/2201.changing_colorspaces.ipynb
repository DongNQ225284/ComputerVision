## Chuyển đổi sang các Không gian màu khác nhau


**`cv2.cvtColor()`** Chuyển đổi một hình ảnh từ không gian màu này sang không gian màu khác. Hàm này chuyển đổi một hình ảnh đầu vào từ không gian màu này sang không gian màu khác. Trong trường hợp chuyển đổi sang không gian màu RGB, thứ tự các kênh phải được chỉ định rõ ràng (RGB hoặc BGR). Lưu ý rằng định dạng màu mặc định trong OpenCV thường được gọi là RGB nhưng thực tế là BGR (các byte bị đảo ngược). Vì vậy, byte đầu tiên trong một hình ảnh màu chuẩn (24 bit) sẽ là thành phần Xanh lam 8 bit, byte thứ hai sẽ là Xanh lục và byte thứ ba sẽ là Đỏ. Sau đó, byte thứ tư, thứ năm và thứ sáu sẽ là pixel thứ hai (Xanh lam, sau đó là Xanh lục, sau đó là Đỏ), v.v.

### Cú pháp hàm 
``` python
dst = cv2.cvtColor( src, code )
```

`dst`: Ảnh đầu ra có cùng kích thước và độ sâu với `src` không.

Hàm có **2 đối số bắt buộc**:

1. Ảnh đầu vào `src`: 8 bit không dấu, 16 bit không dấu ( CV_16UC... ), hoặc dấu phẩy động độ chính xác đơn.

2. Mã chuyển đổi không gian màu `code` (xem ColorConversionCodes).

````python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image

%matplotlib inline
````

````python
img = cv2.imread("images/coca-cola-logo.png", 1)
dst = cv2.cvtColor(img, cv2.COLOR_BGRA2GRAY)
print(dst.shape)

plt.imshow(dst, cmap="gray")
````
