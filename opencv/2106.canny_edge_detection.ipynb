# Phát hiện đặc trưng ảnh

```python
import cv2
import sys
import numpy
import matplotlib.pyplot as plt
import copy
```

### Đọc ảnh

```python
# Đọc ảnh
image_bgr = cv2.imread('image/violet.jpg', cv2.IMREAD_COLOR)
# chuyển sang kênh rgb
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
# Chuyển sang kênh ảnh xám
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
# Thông số đặc trưng
feature_params = dict(maxCorners=500, qualityLevel=0.05, minDistance=30, blockSize=9)
# Phát hiện cạnh
image_canny = cv2.Canny(image_rgb, 140, 150)
# Làm mờ
image_blur  = cv2.blur(image_rgb, (15, 15))
# Phát hiện góc
image_edge  = copy.deepcopy(image_rgb)

corners     = cv2.goodFeaturesToTrack(image_gray, **feature_params)
if corners is not None:
            for x, y in numpy.float32(corners).reshape(-1, 2):
                cv2.circle(image_edge, (int(x), int(y)), 10, (0, 0, 255), 1)

# Hiển thị ảnh
plt.figure(figsize=(9, 9))
plt.subplot(221); plt.imshow(image_rgb);    plt.title("Original")
plt.subplot(222); plt.imshow(image_canny);  plt.title("Canny")
plt.subplot(223); plt.imshow(image_blur);   plt.title("Blur")
plt.subplot(224); plt.imshow(image_edge);   plt.title("Edge")
```
